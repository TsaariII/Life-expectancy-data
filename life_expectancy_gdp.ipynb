```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = pd.read_csv('all_data.csv')
```

The dataset provided has the following columns of data:

Country - nation

Year - the year for the observation

Life expectancy at birth (years) - life expectancy value in years

GDP - Gross Domestic Product in U.S. dollars

```python
print(data.head(15))
```

```python
print(data.Country.unique())
print(data.Year.unique())
```

```python
chile = data[data['Country'] == 'Chile']
print(min(chile['GDP']))
print(max(chile['GDP']))
print(min(chile['Life expectancy at birth (years)']))
print(max(chile['Life expectancy at birth (years)']))

plt.figure(figsize = (22, 8))
plt.subplot(1, 2, 1)
plt.plot(chile['Year'], chile['GDP'])
plt.title('Chile GDP')
ax = plt.subplot(1, 2, 1)
ax.set_xticks(data['Year'])
ax.set_xticklabels(data['Year'])
ax.set_yticks(np.arange(68000000000, 280000000000, 20000000000))

plt.subplot(1, 2, 2)
plt.plot(chile['Year'], chile['Life expectancy at birth (years)'])
plt.title('Chile Life expectancy')
ax = plt.subplot(1, 2, 2)
ax.set_xticks(data['Year'])
ax.set_xticklabels(data['Year'])
ax.set_yticks(np.arange(77, 82, 0.5))
plt.show()
```

Looking into to the life expectancy. Clearly we can see that the life expectancy has increased in every country. Also it is clear thtat in Zimbabwe life expectancy is much lower than in the five other countries.

```python
countries = [chile, china, germany, mexico, usa, zimbabwe]
mean_life = []
for country in countries:
    mean_life.append(np.mean(country['Life expectancy at birth (years)']))

contry_name = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']

plt.figure(figsize=(8,6))
bar_colors = ['tab:red', 'tab:red', 'tab:black', 'tab:forestgreen', 'tab:navy', 'tab:gold']
plt.bar(range(len(mean_life)), mean_life, color=['lightcoral', 'red', 'black', 'forestgreen', 'navy', 'gold'])
ax = plt.subplot()
ax.set_xticks(range(len(contry_name)))
ax.set_xticklabels(contry_name)
plt.title('Life expectancy at birth (years) mean')
plt.ylabel('Age')

plt.figure(figsize=(8,6))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country')

plt.show()
```
